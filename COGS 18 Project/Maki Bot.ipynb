# Project Description

Through this project, I have created an ordering bot of a Chinese restaurant called Makino. This bot's name is Maki. He can welcome and help customers with going over the menu and ordering food.

After greetings, customers can enter "menu" or "categories" to start ordering. Maki is able to display the menu's full version, or to show by categories if customers' comman to do so. Decriptions of each item on the menu would be presented at customers' demand. Customers can freely order or remove food as they think through what would they like to eat for today. Customers can also go back and forth to the menu as cheking different items in detial, and also ask for recommendations from Maki. In the end, after ordering has completed, customers will be directed to the place where they will be informed about the total price of this order.
